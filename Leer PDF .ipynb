```python
import fitz
```

```python
pdf = "pendientes1ss.pdf"
docum = fitz.open(pdf)
```

```python
print("numero de paginas: ",docum.pageCount)
```

```python
pagina = docum.loadPage(0)
text = pagina.getText("text")
```

```python
print(text)
```

```python
len(text)
```

```python
for i in range(len(text)):
    print("Numero: {} caracter {}".format(i,text[i]))
```

```python
import os
salida = open("prueba.txt","wb")

for pag in docum:
    texto =pag.getText().encode("utf8")
    salida.write(texto)
salida.close()
```
